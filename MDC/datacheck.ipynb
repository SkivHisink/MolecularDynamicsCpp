```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('data.csv')
kenetic = np.array(data['kenetic'].to_list())
potential = data['potential'].to_list()
plt.figure(figsize=(16, 14), dpi = 75)
plt.plot(kenetic, label = 'kenetic')
plt.plot(potential, label = 'potential')
plt.plot(potential+kenetic, label = 'summ')
plt.legend()
print(np.min(kenetic))
print(np.min(potential))
print(np.max(kenetic))
print(np.max(potential))
```

```python
data_speed = pd.read_csv('speed.csv')
speed = np.array(data_speed['speed'].to_list())
plt.plot(speed)
print(np.max(speed))
```
